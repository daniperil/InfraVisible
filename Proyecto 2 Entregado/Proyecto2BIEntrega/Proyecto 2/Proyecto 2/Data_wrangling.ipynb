```python
def clean_genero(x):
    
    if not pd.isna(x):
        if 'a.' in x:
            x= x.replace('a. ', '')
        if 'b. ' in x:
            x=x.replace('b. ', '')

        x=x.replace('.', '')

        if 'Mujer' in x:
            x=x.replace('Mujer', 'Femenino')
        if 'Hombre' in x:
            x=x.replace('Hombre', 'Masculino')
        if 'Otro' in x or x == 'reptiliano' or x == 'Gay':
            x='Otro'
    else:
        x = None
        
    return x

df_all_data_m['Género'] = df_all_data_m['Género'].apply(clean_genero)
```

```python
def clean_edad(x):
    
    if not pd.isna(x):
        if x in {'18–30.', '18-30', '18–30, 31–40', '18 - 30', 'a. 18-30', '18–30', '18 a 30 años'}:
            x= '18-30'
        elif x in {'31-40', '31–40.','31–40', '31 a 40 años', '31 - 40', 'b. 31-40', '31-41'}:
            x= '31-40'
        elif x in {'41–50.', '41-50','41–50', '41 a 50 años', '41 - 50', 'c. 41-50'}:
            x= '41-50'
        elif x in {'51–60.', '51-60', '51–60', '51 a 60 años', '51 - 60', 'd. 51-60'}:
            x= '51-60'
        elif x in {'mayor de 60.', 'Más de 60', 'mayor de 60', 'Mayor de 60', 'Mayor de 60 años', 'e. Mayor de 60' }:
            x= 'Mayor de 60'
    else:
        x = None
        
    return x

df_all_data_m['Rango de edad'] = df_all_data_m['Rango de edad'].apply(clean_edad)
```

```python
def clean_ocupacion(x):
    if not pd.isna(x):
        if ('des' not in x.lower())and('empleado' in x.lower() and len(x.split(','))==1):
            x = x.replace('empleado','Empleado')
            x = x.replace('Empleado.','Empleado')
            x = x.replace('a. Empleado','Empleado')
        if ('estudiante' in x.lower() ):
            x = x.replace('estudiante','Estudiante')
            x = x.replace('Estudiante.','Estudiante')
            x = x.replace('Estudiante ','Estudiante')
            x = x.replace('Estudiantey','Estudiante y')
            x = x.replace('b. Estudiante','Estudiante')
        if ('pensionado' in x.lower() ):
            x = x.replace('pensionado','Pensionado')
            x = x.replace('Pensionado.','Pensionado')
            x = x.replace('c. Pensionado','Pensionado')
        if ('independiente' in x.lower() ):
            x = x.replace('independiente','Independiente')
            x = x.replace('Independiente.','Independiente')
            x = x.replace('d. Independiente','Independiente')
        if ('trabajador informal' in x.lower() ):
            x = x.replace('Trabajador Informal.','Trabajador Informal')   
        if ('hogar' in x.lower() ):
            x = x.replace('hogar','Hogar')
            x = x.replace('Hogar.','Hogar')
            x = x.replace('f. Hogar','Hogar')
        if ('desemp' in x.lower() ):
            x = x.replace('Desempleada','Desempleado')
            x = x.replace('Desempleada ','Desempleado')
            x = x.replace('Desempleado ','Desempleado')
            x = x.replace('Desemplead@','Desempleado')
        if ('contratista' in x.lower() ) or 'profesor' in x.lower() or 'docente' in x.lower() or 'programador' in x.lower():
            x = 'Empleado'
        if ('jubilada' in x.lower() ):
            x = x.replace('jubilada','Pensionado')
        if ('un cesante' in x.lower() ):
            x = x.replace('Un Cesante','Desempleado')
        if ('dependiente' in x.lower() ):
            x = x.replace('Dependiente','Desempleado')  
        if ('empresario' in x.lower()  or 'artista' in x.lower()):
            x = x.replace('Empresario','Independiente')
            x = x.replace('Artista','Independiente')
    
        x= x.replace(' y ', ',')
        x= x.replace(', ', ',')
        x= x.replace(' e ', ',')
        
        if 'Empleado,Estudiante' in x or 'Estudiante,Empleado' in x:
            x = x.replace('Empleado,Estudiante','Estudiante,Empleado')
            
        if 'Facebook' in x :
            x = None
        
    else:
        x = None
            
    return x

df_all_data_m['Seleccione su ocupación'] = df_all_data_m['Seleccione su ocupación'].apply(clean_ocupacion)
```
